```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
merged_citrus = pd.read_excel('/Users/zoekim/Desktop/l3/aroma_result/citrus/csv/merged_citrus_v2.xlsx')
merged_citrus = merged_citrus.iloc[:, 1:7]

# Sort by receptor and set receptor as index
merged_citrus_sorted = merged_citrus.sort_values(by='Receptor').set_index('Receptor')

# Generate the clustermap with adjusted parameters
plt.figure(figsize=(15, 50))
clustermap = sns.clustermap(
    merged_citrus_sorted[['Citral', 'Decyl Acetate', 'Limonene', 'Phenylacetic', 'Naringin']],
    annot=False,
    cmap='coolwarm',
    linewidths=.5,
    linecolor='black',
    figsize=(15, 50)
)

# Rotate y-tick labels for better visibility
plt.setp(clustermap.ax_heatmap.yaxis.get_majorticklabels(), rotation=0, fontsize=8)

# Adjust the title position and axis labels
clustermap.fig.suptitle('Clustermap of Affinities Between Receptors and Citrus Ligands Grouped by Odor Characteristics', y=1.05)
clustermap.ax_heatmap.set_xlabel('Ligands')
clustermap.ax_heatmap.set_ylabel('Receptors')
plt.savefig("clustermap_citrus_v1.png", format="png")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
merged_citrus = pd.read_excel('/Users/zoekim/Desktop/l3/aroma_result/citrus/csv/merged_citrus_v2.xlsx')
merged_citrus = merged_citrus.iloc[:, 1:7]

# Normalize the data
normalized_citrus = merged_citrus.copy()
columns_to_normalize = ['Citral', 'Decyl Acetate', 'Limonene', 'Phenylacetic', 'Naringin']

for column in columns_to_normalize:
    normalized_citrus[column] = (merged_citrus[column] - merged_citrus[column].min()) / (merged_citrus[column].max() - merged_citrus[column].min())

# Sort by receptor and set receptor as index
normalized_citrus_sorted = normalized_citrus.sort_values(by='Receptor').set_index('Receptor')

# Generate the clustermap with adjusted parameters
plt.figure(figsize=(15, 50))
clustermap = sns.clustermap(
    normalized_citrus_sorted[['Citral', 'Decyl Acetate', 'Limonene', 'Phenylacetic', 'Naringin']],
    annot=False,
    cmap='coolwarm',
    linewidths=.5,
    linecolor='black',
    figsize=(15, 50)
)

# Rotate y-tick labels for better visibility
plt.setp(clustermap.ax_heatmap.yaxis.get_majorticklabels(), rotation=0, fontsize=8)

# Adjust the title position and axis labels
clustermap.fig.suptitle('Clustermap of Affinities Between Receptors and Citrus Ligands Grouped by Odor Characteristics (Normalized)', y=1.05)
clustermap.ax_heatmap.set_xlabel('Ligands')
clustermap.ax_heatmap.set_ylabel('Receptors')

# Save the clustermap
plt.savefig("clustermap_citrus_normalized_v1.png", format="png")
plt.show()
```
